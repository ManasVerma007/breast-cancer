### Categorical variable encoding

```python
data["diagnosis"].value_counts() # so the data is almost balanced
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data["diagnosis"] = encoder.fit_transform(data["diagnosis"])
data.head()
```

```python
data["diagnosis"].value_counts()
```

```python
X = data.drop(columns='diagnosis', axis=1)
y = data["diagnosis"]
```

```python
X.head()
```

# Drop the constant feature

```python
### so here i will drop the columns that has constant values
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.05)
selector.fit(X)
selector.get_support()
```

```python
## so here i will check how many feature is remain
sum(selector.get_support())
```

```python
X.columns[selector.get_support()]
```

```python
X = X[X.columns[selector.get_support()]]
```

```python
X.head()
```

## Split the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.72, random_state=42)
```

```python
X_train.shape
```

### Here i will do the feature scaling and also create the pipeline

```python
## so here i will use the pipeline to do the same thing
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression 

steps = [('scaler', StandardScaler()), ('classifier', LogisticRegression())]
pipeline = Pipeline(steps)
```

## Train and select the model

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
## so here we check the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Here we build the model 

```python
import pickle
pickle.dump(pipeline, open("model.pkl", "wb"))
```
